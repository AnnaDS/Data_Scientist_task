```python
import pandas as pd
import datetime
df= pd.read_csv('/Users/annachystiakova/Documents/Oracle_Data scientist job_test task/logs.csv', sep='\t',header=None)
df=df[[0,1,2,9]]

def to_datet(x):
    return datetime.datetime.fromtimestamp(int(str(x))).strftime('%Y-%m-%d %H:%M:%S')

df[0]=[to_datet(x) for x in df[0]]
df.index=df[0]
df=df[[1,2,9]]
df.columns=['host', 'process','data']
df=df[df['process'].str.contains("App_Srv process")]
df = df.dropna()#[np.isfinite(df['data'])]
df=df[df['data'].str.contains("vsz=")]
df.head()
```

```python
import numpy as np
def get_vsz(x):
    x=x.split(";;")
    #print (len(x))
    if len(x)<2:
        return NaN
    else:
        x=x[1].split(";")[0]
        return int(x.replace("vsz=",""))

df['data']=[get_vsz(x) for x in df['data'].values]
```

```python
print("The number of records in result table with host and App-Srv process is "+str(df.shape[0]))
print("Head of result table ")
df.head()
```

```python
groups=df.groupby(df['host'])['process'].unique()
groups.to_csv('/Users/annachystiakova/Documents/Oracle_Data scientist job_test task/output_1.csv')
```
